# Section 1.1: Introduction

Solving linear equations involves analysis of _$\bf{n}$ equations in $\bf{n}$ unknowns_.
* _Gaussian elimination_: Multiples of the first equation are subtracted from other equations
* _Cramer's Rule_: Solves as a ratio of two $\bf{n} \times \bf{n}$ determinants


# Section 1.2: The Geometry of Linear Equations

Consider a system of equations:

$$2x-y=1\\
x+y=5$$

The classic approach looks at the _rows_ in the problem ($2x-y=1$ and $x+y=5$ separately). The second approach, though, uses _columns_, looking at the two equations as a single _**vector equation**_:

$$x \begin{bmatrix} 2\\1 \end{bmatrix} + y \begin{bmatrix}-1\\1 \end{bmatrix} = \begin{bmatrix} 1\\5 \end{bmatrix}$$

The objective is to find the combination of column vectors on the left to produce the vector on the right.

For example, in Figure 1.1, Figure (A) represents the two linear equations, and Figure (B) represents the vector equations, where the first vector is green and the second vector is purple:

```python
figures.figure_1_1()
```

In Figure 1.1(B), the objective is to discover what $x$ and $y$ are required to transform the original vectors to the new vector. Visually, we can see that $x=2$ and $y=3$.

Note that with more equations the principle remains the same, even as visualization falls apart. For example, in three dimensions, any given equation represents a _plane_ rather than a _line_. (All planes represent $n-1$ dimensions. Equations are therefore "solved" at a single point, whose coordinates intersect all dimensions.)

### Column Vectors

Consider the following system equations:

$$\begin{matrix} 2u & + &  v & + &  w & = &  5\\
                 4u & - & 6v &   &    & = & -2\\
                -2u & + & 7v & + & 2w & = &  9\end{matrix}$$
                
This system can be rewritten as the vector equation:

$$u \begin{bmatrix} 2\\4\\-2 \end{bmatrix} + v\begin{bmatrix} 1\\-6\\7 \end{bmatrix} + w\begin{bmatrix} 1\\0\\2 \end{bmatrix} = \begin{bmatrix} 5\\-2\\9 \end{bmatrix}$$

The righthand vector (let's call it $b$) has components $5,-2,9$ and is identified with the point located that distance from the origin through three dimensions.

_**Vector addition**_ represents vector $b$ as follows:

$$\begin{bmatrix}5\\0\\0\end{bmatrix} + \begin{bmatrix}0\\-2\\0\end{bmatrix} + \begin{bmatrix}0\\0\\9\end{bmatrix} = \begin{bmatrix}5\\-2\\9\end{bmatrix}$$

_**Scalar multiplication**_ involves multiplying by a scalar (a single number which _scales_ the vector):

$$2\begin{bmatrix}1\\0\\2\end{bmatrix} = \begin{bmatrix}2\\0\\4\end{bmatrix}$$

A _**linear combination**_ is then a combination of multiple methods of vector arithmetic, and our equation above is solved with

$$\begin{bmatrix}2\\4\\-2\end{bmatrix} + \begin{bmatrix}1\\-6\\7\end{bmatrix} + 2\begin{bmatrix}1\\0\\2\end{bmatrix} = \begin{bmatrix}5\\-2\\9\end{bmatrix}$$

where $\begin{bmatrix}1&1&2\end{bmatrix}$ is the _linear combination_ of multipliers $u$, $v$, and $w$.

Some definitions, then:
* _**Row picture:**_ Intersection of $n$ planes
* _**Column picture:**_ The right side $b$ is a combination of the column vectors
* _**Solution to equations:**_ Intersection point of planes = coefficients in the combination of columns

# Section 1.3: An Example of Gaussian Elimination

Consider the following system:
$$\begin{matrix}
 2u &+&  v &+&  w &=&  5\\
 4u &-& 6v & &    &=& -2\\
-2u &+& 7v &+& 2w &=&  9
\end{matrix}$$

The objective of the elimination is to isolate each of the values $u$, $v$, and $w$. To do so, examine the _**pivot**_ of each row--the diagonal's coefficient--and use multiples of that pivot to eliminate the coefficients from the rows below that pivot. (Note: Pivots cannot be zero. If a zero appears, the solution may be singular.)

**Step 1:** _Subtract multiples of the first equation from the others, so as to eliminate $u$ from the last two equations._
1. Subtract $2\times$ the first equation from the second;
2. Subtract $-2\times$ the first equation from the third.

The result:
$$\begin{matrix}
2u &+&   v &+&  w &=&   5\\
   & & -8v &-& 2w &=& -12\\
   & &  8v &+& 3w &=&  14
\end{matrix}$$

**Step 2:** _Subtract a multiple of the second equation from the third, so as to eliminate $v$ from the final equation._ 
1. Subtract $-1$ times the second equation from the third.

The result:
$$\begin{matrix}
2u &+&  v &+&  w &=&   5\\
   &-& 8v &-& 2w &=& -12\\
   & &    & &  w &=&   2
\end{matrix}$$

Now that we have a basic substitution $w=2$, we can perform _**back-substitution**_ to find $v=1$ and $u=1$.

In matrix form, include the solution as part of the matrix:

$$\begin{bmatrix}2&1&1&5\\4&-6&0&-2\\-2&7&2&9\end{bmatrix}
\Rightarrow
\begin{bmatrix}2&1&1&5\\0&-8&-2&-12\\0&8&3&14\end{bmatrix}
\Rightarrow
\begin{bmatrix}2&1&1&5\\0&-8&-2&-12\\0&0&1&2\end{bmatrix}$$

This is known as the _triangular_ system. If a system can reach the triangular system, it has a single solution.

If a zero pivot is encountered, sometimes the equation simply needs to be re-ordered:

$$\begin{bmatrix}1&1&1\\2&2&5\\4&6&8\end{bmatrix}
\Rightarrow
\begin{bmatrix}1&1&1\\0&0&3\\0&2&4\end{bmatrix}
\Rightarrow
\begin{bmatrix}1&1&1\\0&2&4\\0&0&3\end{bmatrix}$$
